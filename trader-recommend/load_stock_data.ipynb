```python
def _parse_dates(params, date_from, date_to):
    # start of request period. Default is 1-1-1990
    date_1 = parse('1-1-1990')
    try:
        date_1 = parse(date_from)        
    except:
        print 'Incorrect from date'
    finally:
        params['d1'] = date_1.day
        params['m1'] = date_1.month
        params['y1'] = date_1.year

    # end of request period. Default is today date
    date_2 = datetime.datetime.now()
    try:
        date_2 = parse(date_to)
    except:
        print 'Date from Exception'
    finally:
        params['d2'] = date_2.day
        params['m2'] = date_2.month
        params['y2'] = date_2.year
    
    if (date_2.date() == datetime.datetime.now().date()):
        params['lastdays'] = date_2.date()-date_1.date()
    else:
        params['lastdays'] = 'NULL'
    
def get_stock_data(ticker, period = "DAILY", date_from = None, date_to = None):
    host = "http://export.rbc.ru/free/micex.0/free.fcgi"
    params = {}
    params['period'] = period
    # short name for finance entity. IF null server will return all the data
    params['tickers'] = ticker
    
    _parse_dates(params, date_from, date_to)
    
    # number of days. It's delta between end and start period. 
    # It should be NULL if end of request period is not today
    
    params['separator'] = "TAB"
    # 
    params['data_format'] = "BROWSER"
    # get header for data. 1 means enable, 0 - otherwise
    params['header'] = "1"
    
    url_values = urllib.urlencode(params)
    url = host +'?'+ url_values
    print url
    return pd.read_csv(url, delim_whitespace=True, header=0, parse_dates=['DATE'])
```
